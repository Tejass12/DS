```python
Q1=df.Math_score.quantile(0.25)
```

```python
Q1
```

```python
Q3=df.Math_score.quantile(0.75)
```

```python
Q3
```

```python
IQR=Q3-Q1
```

```python
IQR
```

```python
lowerlimit=Q1-1.5*IQR
```

```python
upperlimit=Q3+1.5*IQR
```

```python
lowerlimit
```

```python
upperlimit
```

```python
outlier=df["Math_score"][(df["Math_score"]>upperlimit)|(df["Math_score"]<lowerlimit)]
```

```python
outlier
```

```python
mean1=df["Math_score"].mean()
```

```python
df["Math_score"]=np.where((df["Math_score"]>upperlimit),mean1,df["Math_score"])
```

```python
mean1=df["Reading_score"].mean()
```

```python
df["Reading_score"]=np.where((df["Reading_score"]>upperlimit),mean1,df["Reading_score"])
```

```python
mean1=df["Writing_score"].mean()
```

```python
df["Writing_score"]=np.where((df["Writing_score"]>upperlimit),mean1,df["Writing_score"])
```
